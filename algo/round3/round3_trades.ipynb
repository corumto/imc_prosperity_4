```python
import pandas as pd
from pathlib import Path

base = Path(r"c:/Users/tolan/OneDrive/MDS Stuff/IMC/round3")
files = [base / f"trades_round_3_day_{day}.csv" for day in range(3)]

rows = pd.concat((pd.read_csv(file, sep=";") for file in files), ignore_index=True)
summary = (
    rows.groupby("symbol", as_index=False)
    .agg(trades=("symbol", "size"), total_quantity=("quantity", "sum"))
    .sort_values("symbol")
)

summary
```
