```python
import pandas as pd
import matplotlib.pyplot as plt

# Load both CSV files
df1 = pd.read_csv('prices_round_1_day_0_test.csv', sep=';')
df2 = pd.read_csv('prices_round_2_day_0_test.csv', sep=';')

# Drop NA and 0 values in mid_price
df1 = df1.dropna(subset=['mid_price'])
df1 = df1[df1['mid_price'] != 0]
df2 = df2.dropna(subset=['mid_price'])
df2 = df2[df2['mid_price'] != 0]

# Create figure with subplots for each asset
fig, axes = plt.subplots(2, 1, figsize=(12, 8))

# Plot ASH_COATED_OSMIUM
ash1 = df1[df1['product'] == 'ASH_COATED_OSMIUM'].sort_values('timestamp')
ash2 = df2[df2['product'] == 'ASH_COATED_OSMIUM'].sort_values('timestamp')

axes[0].plot(ash1['timestamp'], ash1['mid_price'], marker='o', label='Round 1', linewidth=2)
axes[0].plot(ash2['timestamp'], ash2['mid_price'], marker='s', label='Round 2', linewidth=2)
axes[0].set_xlabel('Timestamp')
axes[0].set_ylabel('Mid Price')
axes[0].set_title('ASH_COATED_OSMIUM - Mid Price Comparison')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Plot INTARIAN_PEPPER_ROOT
root1 = df1[df1['product'] == 'INTARIAN_PEPPER_ROOT'].sort_values('timestamp')
root2 = df2[df2['product'] == 'INTARIAN_PEPPER_ROOT'].sort_values('timestamp')

axes[1].plot(root1['timestamp'], root1['mid_price'], marker='o', label='Round 1', linewidth=2)
axes[1].plot(root2['timestamp'], root2['mid_price'], marker='s', label='Round 2', linewidth=2)
axes[1].set_xlabel('Timestamp')
axes[1].set_ylabel('Mid Price')
axes[1].set_title('INTARIAN_PEPPER_ROOT - Mid Price Comparison')
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
